Now we will implicitly define a radial light source

```python
class LightSource:
    def __init__(self):
        self.origin = np.array([250, 250])
        self.radius = 250

    def _query(self, coordinate, normals, alpha=0.005):
        direction = coordinate - self.origin
        euclidean_distance = npl.norm(self.origin - coordinate)
        new_shading = normals * direction
        new_shading = np.sum(new_shading, axis=2)

        return 1.0 / (1 + alpha * euclidean_distance)

    def draw_image(self, image, normals):
        image_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
        for i, _ in enumerate(image_hsv[0]):
            for j, _ in enumerate(image_hsv[1]):
                light_v = self._query(np.array([i, j]), normals)
                image_hsv[i, j, 2] += light_v
        image_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)
        return image_rgb
```

```python
light = LightSource()

image_relit = light.draw_image(image)

fig, ax = plt.subplot_mosaic([['image', 'image_relit']], figsize=(7, 3.5))
ax['image'].imshow(image)
ax['image'].axis('off')
ax['image'].set_title('Image')

ax['image_relit'].imshow(image_relit)
ax['image_relit'].axis('off')
ax['image_relit'].set_title('Relit Image')

markers = [(250, 250)]
x, y = zip(*markers)
# ax['image_relit'].plot(x, y, 'o')

plt.show()
```

## Recreate the original image from albedo and shading

```python
albedo = plt.imread('./church/ours_alb.png')
shading = plt.imread('./church/ours_shd.png')
shading = np.dstack([shading, shading, shading])

lin_recreate = albedo * shading
gamma_correct = lin_recreate ** 0.4545
plt.imshow(gamma_correct)
```

Create direction vector field for a given light position

```python
depth = plt.imread('./church/depth_metric_r20.png')
origin = [250, 250, depth[250, 250]]

h, w = gamma_correct.shape[0], gamma_correct.shape[1]
N = h * w

depth_vector_field = np.zeros_like(gamma_correct)
depth_vector_field = np.reshape(depth_vector_field, [N, -1])
depth_vector_field[:, :, 0] = np.arange(0, w)
depth_vector_field[:, :, 1] = np.arange(0, h)
depth_vector_field[:, :, 2] = depth
print(f"{depth_vector_field.shape= }")

depth_vectorized = np.reshape(depth, [N, -1])
print(depth_vectorized.shape)
print(f"{origin= }, {h= }*{w= }={N}")
```
